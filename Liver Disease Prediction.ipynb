# encoding

 When we are doing ML model we required library called sklearn.For the preparing data we used package called preprocessing.
 Then we import two classes labelencoder(for converting text to numbers),ordinalencder(for converting categorical columns in tables into 2 numbers).
 then we created an object of labelencoder
 Here we used a function(encoder.fit_transform) means .fit is used for reading text values and transform is used for converting it into numeric values . 

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
encoder=LabelEncoder()
df['Gender']=encoder.fit_transform(df['Gender'])
```

# Feature Scaling: scaling is used for to ensures all columns numerical value on a similar range for better model performance because some columns values has higher so it will not take as a similar value.


```python
# In this we used class standardscaler for making all the columns in similar range
```

```python
from sklearn.preprocessing import StandardScaler
features=['Age','Total_Bilirubin','Direct_Bilirubin',
          'Alkaline_Phosphotase','Alamine_Aminotransferase',
          'Aspartate_Aminotransferase','Total_Protiens','Albumin',
          'Albumin_and_Globulin_Ratio']
scaler=StandardScaler()
df[features]=scaler.fit_transform(df[features])
df
          
```

# Data Balancing:In this we make the balance of count of disease patient and count of healthy patient 

```python
pip install -q imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
X=df.drop('Dataset',axis=1)
y=df['Dataset']
smote=SMOTE(random_state=42)
X_resampled,y_resampled=smote.fit_resample(X,y)
y_resampled.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_resampled,y_resampled,test_size=0.2,random_state=42)
```

# Train model-Logistic regression,svc,decision tree,random forest

# Logistic regression-a specific machine learning algorithm used for classification (predicting categories like Yes/No, 0/1, etc

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
model=LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("\n\n",classification_report(y_test,y_pred))
print("\n\n",confusion_matrix(y_test,y_pred))
```

# Randomforestclassifier-RandomForestClassifier is a machine learning algorithm from the ensemble family that uses multiple decision trees to perform classification tasks.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,classification_report
model_r=RandomForestClassifier()
model_r.fit(X_train,y_train)
y_pred=model_r.predict(X_test)
print("\n\n",classification_report(y_test,y_pred))
print("\n\n",confusion_matrix(y_test,y_pred))
```

# Support vector classifier-SVC is like a line-drawer it draws the best possible line (or curve) between different categories of data.

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report
model=SVC()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("\n\n",classification_report(y_test,y_pred))
print("\n\n",confusion_matrix(y_test,y_pred))
```
